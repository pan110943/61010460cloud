```
customer = [('win','0618201998'),('ricky','08172312313'),('woon','09123144513'),('somchai','0712314534')]
n = 10
for i in range(n): #loop เพื่อใส่ชื่อ win กับ tel n ครั้ง
  customer.append(('winn','01231412314'))
customer.append(('mike','018512314123')) #เพื่อชื่อ mike อยู่ในตำแหน่งสุดท้ายของตัวแปร customer
```

```
def getphonenum(cusname):
  for c in customer: #loop ทุกชื่อและเบอ 
    #print(c)
    if c[0]==cusname: #c[0] คือชื่อ ถ้าเท่ากับ ชื่อต้องการค้นหา
      cnum = c[1] #ให้นำ c[1] ก็คือเบอออก
      return cnum
```

ใช้เมื่อไหร่

  ใช้หาข้อมูล เมื่อต้องการค้นหา เช่น เบอร์ อีเมล

```
def creatdatabaselist(n):
  listdb = [] #ประกาศว่าสร้าง list
  for i in range(n): #วนลูป n ครั้ง
    listdb.append(('pan'+str(i),i+100)) #ใส่ข้อมูลเป็น tuple () ใส่ str(n) ใส่ str(n) เป็น element ตัวแรก และ i + 100 เป็น element
  return listdb

n = 10000 #สร้างข้อมูลจำลอง n record
listdb = creatdatabaselist(n)
listdb
```

```
def creatdatabasedict(n):
  dictdb = {} #ประกาศว่าจะสร้าง dic
  for i in range(n): #วนลูป n ครั้ง
    dictdb['pan'+str(i)] = i + 100 #ใส่ str(n) เป็น key และใส่ i + 100 เป็นค่า
  return dictdb
n = 10000 #สร้างข้อมูลจำลอง n record
dictdb = creatdatabasedict(n)
dictdb
```

```
def getphonenum(cusname,listdb):
  for c in listdb:
    if c[0]==cusname:
      cnum = c[1]
  return cnum
```

```
getphonenum('pan3',listdb)
```

```
%time getphonenum('pan9999',listdb) #จับเวลาาา
```
